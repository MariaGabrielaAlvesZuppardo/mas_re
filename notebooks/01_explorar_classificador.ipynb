## Explorando o  Agente de Classificação de Requisitos

```python
import os
from dotenv import load_dotenv

os.chdir("..")
load_dotenv()
```

```python
from datasets.promise import PromiseAdapter
from pipeline.state import Requirement
from agents.classificador import ClassificationAgent
```

```python
adapter = PromiseAdapter()
data = adapter.load(sample_n=10,random_state=42)

requirements = [
    Requirement(
        id=f"PROMISE-{item.project_id}-{i:03d}",
        text=item.requirement,
        source=f"promise_nfr/project_{item.project_id}",
    )
    for i, item in enumerate(data.items, 1)
]
```

```python
agente = ClassificationAgent(model="llama-3.3-70b-versatile")

for req in requirements:
    resultado = agente.classify(req)
    print(f"{resultado.id} | {resultado.category.value} | {resultado.subcategory} | {resultado.confidence:.2f}")
    print(f"  {resultado.text[:80]}")
    print()
```

### Explorando o dataset em pt-br

```python
import pandas as pd
```

```python
df = pd.read_csv("datasets/promise_nfr/Promise+_pt.csv")
print(f"Total: {len(df)} requisitos")
print(f"FR:  {(df['label'] == 'FR').sum()}")
print(f"NFR: {(df['label'] == 'NFR').sum()}")
df.head()
```

```python
requirements = [
    Requirement(
        id=f"PROMISE-{row['project_id']}-{i:03d}",
        text=row["requirement_pt"],
        source=row["source"],
    )
    for i, row in df.head(10).iterrows()
]

for req in requirements:
    print(f"{req.id} | {req.text[:80]}")
```

```python
agente = ClassificationAgent(model="llama-3.3-70b-versatile")

for req in requirements:
    resultado = agente.classify(req)
    print(f"{resultado.id} | {resultado.category.value} | {resultado.subcategory} | conf: {resultado.confidence:.2f}")
    print(f"  {resultado.text[:80]}")
    print()
```
